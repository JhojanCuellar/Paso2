```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("D:/Desktop/data.csv", sep=",")
```

```python
data
```

```python
data.plot.scatter(x="precio", y="metro", color="orange")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
precio = data["precio"].values.reshape((-1, 1))
```

```python
model = regresion.fit(precio, data["metro"])

print("Interseccion (b)", model.intercept_)
print("Pendiente (m)", model.coef_)
```

```python
entrada = [[200], [420], [490]]
predicciones = model.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="precio", y="metro", label='Primarios')
plt.scatter(entrada, predicciones, color='orange')
plt.plot(entrada, predicciones, color='black', label='Linea de regresion')
plt.xlabel('precio')
plt.ylabel('metros')
plt.legend()
plt.show()
```
